```python
from google.colab import files
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
```

```python
image_url = 'https://tinyurl.com/5c5bpu8t'
```

```python
with urllib.request.urlopen(image_url) as response:
    img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)
```

```python
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def show(img,title):
    plt.imshow(img)
    plt.axis('off')
    plt.title(title)
    plt.show()
```

```python
show(img_rgb,'Original Image')
```

a. Resize an image to 256x256 pixels.


```python
img_resize = cv2.resize(img_rgb, (256, 256))
show(img_resize,'Image resize')
```

b. Rotate by 180 degrees clockwise.

```python
img_ro = cv2.rotate(img_rgb, cv2.ROTATE_90_CLOCKWISE)
img_rotate = cv2.rotate(img_ro, cv2.ROTATE_90_CLOCKWISE)
show(img_rotate,'rotate')
```

c. Apply a vertical flip.

```python
img_flip_vertical = cv2.flip(img_rgb, 0)
show(img_flip_vertical,'flip vertical')
```

d. Display all transformations in a single subplot.


```python
plt.subplot(1, 4, 1)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(img_resize)
plt.title('Image resize')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(img_rotate)
plt.title('Image rotate')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(img_flip_vertical)
plt.title('Image flip vertical')
plt.axis('off')

plt.show()
```
